```python
fnames = '/content/colored_images'
```

```python
retinal = DataBlock(blocks = (ImageBlock, CategoryBlock),
                    get_items = get_image_files,
                    get_y = parent_label,
                    splitter = RandomSplitter(valid_pct = 0.2, seed=42),
                    item_tfms = Resize(460),
                    batch_tfms = aug_transforms(size=224, min_scale=0.75))
dls = retinal.dataloaders(fnames, bs=64)
```

```python
dls.show_batch(nrows=2, ncols=2)
```

```python
metrics = [accuracy, RocAuc(average='macro')]
learn_dense201 = vision_learner(dls, densenet201, metrics=metrics)
```

```python
learn_dense201.fine_tune(17)
```

```python
metrics = [accuracy, RocAuc(average='macro')]
learn_vgg = vision_learner(dls, vgg16, metrics=metrics)
```

```python
learn_vgg.fine_tune(15)
```

```python
interp = ClassificationInterpretation.from_learner(learn_vgg)
interp.plot_confusion_matrix()
```

```python
learn_vgg.export('retinal_disease_vgg16.pkl')
```
